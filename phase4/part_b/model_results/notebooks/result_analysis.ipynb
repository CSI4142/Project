# Model Result Comparison Analysis

Here we will compare the results of different models to see how each performed under our dataset. 

### Import matplotlib and numpy to do create graphs with Python.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os

# directory where the results are stored
results_dir = "../"
```

### Define variables based on data results

```python
# list to store the dataframes
df_results = []
# for each file in the results directory, read the csv file and append it to the df_results list
for file in os.listdir(results_dir):
    if file.endswith(".csv") and not file.startswith("["):
        df_results.append(pd.read_csv(results_dir + file))
# concatenate all the dataframes in the df_results list
df_results = pd.concat(df_results)
# sort results by Accuracy
df_results = df_results.sort_values(by="Accuracy", ascending=False)
df_results
```

### Create and Display Graph to Show Results

```python
# the label locations
x = np.arange(len(df_results["Model"]))  

 # the width of the bars
width = 0.15

fig, ax = plt.subplots()
rects1 = ax.bar(x - 2*width, df_results["Accuracy"], width, label='Accuracy')
rects2 = ax.bar(x - width, df_results["Precision"], width, label='Precision')
rects3 = ax.bar(x, df_results["Recall"], width, label='Recall')
rects4 = ax.bar(x + width, df_results["F1 score"], width, label='F1 Score')
rects5 = ax.bar(x + 2*width, df_results["Training Time"], width, label='Training Time')
rects6 = ax.bar(x + 3*width, df_results["Prediction Time"], width, label='Prediction Time')

# graph labels and markings
ax.set_ylabel('Scores')
ax.set_title('Model Comparison')
ax.set_xticks(x)
ax.set_xticklabels(df_results["Model"])
ax.legend()

plt.show()
```

### Create a Line Graph

```python
plt.figure(figsize=(10, 6))
plt.plot(df_results["Model"], df_results["Accuracy"], marker='o', label='Accuracy')
plt.plot(df_results["Model"], df_results["Precision"], marker='o', label='Precision')
plt.plot(df_results["Model"], df_results["Recall"], marker='o', label='Recall')
plt.plot(df_results["Model"], df_results["F1 score"], marker='o', label='F Score')
plt.plot(df_results["Model"], df_results["Training Time"], marker='o', label='Training Time')
plt.plot(df_results["Model"], df_results["Prediction Time"], marker='o', label='Prediction Time')

plt.xlabel('Models')
plt.ylabel('Scores')
plt.title('Model Comparison Line Graph')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
