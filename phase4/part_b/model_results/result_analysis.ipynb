# Model Result Comparison Analysis

Here we will compare the results of different models to see how each performed under our dataset. 

### Import matplotlib and numpy to do create graphs with Python.

```python
import matplotlib.pyplot as plt
import numpy as np
```

### Define variables based on data results

```python
models = ['Random Forest', 'Decision Tree', 'Gradient Boosting']
accuracy = [0.998, 0.996333, 0.998333]
precision = [0.9934959349593496, 0.988836, 0.996689]
recall = [0.9967373572593801, 0.99359, 0.995041]
training_time = [0.1495068073272705, 0.01139, 0.704462]
f_score = [0.995114006514658, 0.991207, 0.995864]
```

### Create and Display Graph to Show Results

```python
# the label locations
x = np.arange(len(models))  

 # the width of the bars
width = 0.15

fig, ax = plt.subplots()
rects1 = ax.bar(x - 2*width, accuracy, width, label='Accuracy')
rects2 = ax.bar(x - width, precision, width, label='Precision')
rects3 = ax.bar(x, recall, width, label='Recall')
rects4 = ax.bar(x + width, training_time, width, label='Training Time')
rects5 = ax.bar(x + 2*width, f_score, width, label='F1 Score')

# graph labels and markings
ax.set_ylabel('Scores')
ax.set_title('Model Comparison')
ax.set_xticks(x)
ax.set_xticklabels(models)
ax.legend()

plt.show()
```

### Create a Line Graph

```python
plt.figure(figsize=(10, 6))
plt.plot(models, accuracy, marker='o', label='Accuracy')
plt.plot(models, precision, marker='o', label='Precision')
plt.plot(models, recall, marker='o', label='Recall')
plt.plot(models, training_time, marker='o', label='Training Time')
plt.plot(models, f_score, marker='o', label='F Score')

plt.xlabel('Models')
plt.ylabel('Scores')
plt.title('Model Comparison Line Graph')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
