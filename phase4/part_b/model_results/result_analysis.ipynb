# Model Result Comparison Analysis

Here we will compare the results of different models to see how each performed under our dataset. 

### Import matplotlib and numpy to do create graphs with Python.

```python
import matplotlib.pyplot as plt
import numpy as np
```

### Define variables based on data results

```python
models = ['Random Forest', 'Decision Tree', 'Gradient Boosting']
accuracy = [0.9986666666666668, 0.998, 0.998]
precision = [0.9933110367892976, 0.9983498349834984, 0.990923]
recall = [1.0, 0.9918032786885246, 1.0]
training_time = [0.159419298171997, 0.008754253387451172, 0.494657]
```

### Create and Display Graph to Show Results

```python
# the label locations
x = np.arange(len(models))  

 # the width of the bars
width = 0.2 

fig, ax = plt.subplots()
rects1 = ax.bar(x - width, accuracy, width, label='Accuracy')
rects2 = ax.bar(x, precision, width, label='Precision')
rects3 = ax.bar(x + width, recall, width, label='Recall')
rects4 = ax.bar(x + 2*width, training_time, width, label='Training Time')

# graph labels and markings
ax.set_ylabel('Scores')
ax.set_title('Model Comparison')
ax.set_xticks(x)
ax.set_xticklabels(models)
ax.legend()

plt.show()
```
