# Data Summarization

In this step, we are trying to analyze the data and get the "feel" of the data. In order to do that we look at the data values, columns, and distribution. We also create plots of graphs to see if there are any obvious trends present in the data. This step is necessary for the next steps if we want to train a machine learning model to analyze the data and predict customer churn.

In the step below, we import the necessary libraries such as *matplotlib* for creating different plots such as scatter plots, histograms, and pie charts. We also print the first few rows of data to view the data values.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("../data/customer_churn_unprocessed.csv")

# Display the first few rows of the dataset
print("First few rows of the dataset:")
print(df.head())
```

```python
# Check for missing values
print("\nMissing values in the dataset:")
print(df.isnull().sum())
```

In order to get the "feel" of data, we will create different plots and graphs to see the distribution and trends in the data. 

```python
# Data summarization
plt.figure(figsize=(12, 6))

# Histogram
plt.subplot(1, 3, 3)
sns.histplot(data=df, x='Satisfaction Score', hue='Exited')

plt.tight_layout()
plt.show()
```

```python
creditscore = df['CreditScore']
points = df['Point Earned']

# Plotting the graph
plt.figure(figsize=(8, 6))
plt.scatter(creditscore, points, color='blue', alpha=0.5)
plt.title('Relation between Credit Score and Points Earned')
plt.xlabel('Credit Score')
plt.ylabel('Points Earned')
plt.grid(True)
plt.show()
```

We can see the relationship between Credit Score and Points Earned in the graph above. We can see that with higher Credit Score, the probability of Points Earned is also higher. 

```python
plt.figure(figsize=(12, 6))

# Histogram
plt.subplot(1, 3, 3)
sns.histplot(data=df, x='Satisfaction Score', y='CreditScore', hue='Exited')

plt.tight_layout()
plt.show()
```

We can see from the above histogram that the customer churn is present at all levels of Satisfaction score and all levels of Credit score. We see that concentration of customers exiting the bank is almost the same at all satisfaction score values.

```python
# Extracting credit score and satisfaction score from the DataFrame
age = df['Age']
balance = df['Balance']

plt.figure(figsize=(8, 6))
plt.scatter(age, balance, color='yellow', alpha=0.5)
plt.title('Relation between Age and Balance of Customer')
plt.xlabel('Age')
plt.ylabel('Balance')
plt.grid(True)
plt.show()
```

We can see from the above scatter plot that there are customers of all ages that have zero balance. However, most of the customer balance is concentrated between 60000 to 170000 between the age of 22 to 60.

```python
# Calculate the proportion of exited customers
exited_count = df['Exited'].sum()  # 'exited' is a binary column indicating whether the customer has exited (1) or not (0)
total_customers = len(df)
exited_percentage = (exited_count / total_customers) * 100

# Labels and sizes for the pie chart
labels = ['Exited', 'Active']
sizes = [exited_percentage, 100 - exited_percentage]

# Create the pie chart
plt.figure(figsize=(8, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=['red', 'yellow'])
plt.title('Percentage of Customer Churn')
plt.axis('equal')
plt.show()
```

```python
exited_count = df['Complain'].sum()  # 'Complain' is a binary column indicating whether the customer has complained (1) or not (0)
total_customers = len(df)
exited_percentage = (exited_count / total_customers) * 100

# Labels and sizes for the pie chart
labels = ['Complain', 'No complain']
sizes = [exited_percentage, 100 - exited_percentage]

# Create the pie chart
plt.figure(figsize=(8, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=['red', 'yellow'])
plt.title('Percentage of Customer Complains')
plt.axis('equal')
plt.show()
```

From the two pie charts show above for the percentage of customer complains and customer churn, we can deduce that the customers who complained exited the company. This helps the bank to make valuable decisions for the future and how to better deal with customer complains in order to retain customers.
